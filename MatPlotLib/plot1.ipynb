Data Visualization

Data Visualization ka matlab hai data ko aise dikhana ki aankhon se dekhte hi samajh aa jaye.
Jaise boring numbers ko graph, chart, ya picture bana ke dikhana

MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]  #X-Axis
```

```python
y = [10,20,15,25]  #Y-Axis
```

For Plotting we will need X and Y 

plot() function is a reusable tool, gives a plotted line under x and y axis

```python
plt.plot(x, y)  #left side mai Horizontal(x-axis) and right side mai vertical(y-axis)

#OR

plt.show()
```

1 = 10
