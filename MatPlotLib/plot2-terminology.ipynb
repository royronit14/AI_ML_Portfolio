<h1>Terminologies used in MATPLOTLIB </h1>

<h2>plt.show()</h2>

Kya hai?

Final command — chart screen par dikhata hai.

Layman:

"Ab tasveer bana ke dikha do!"

<h2> plt.plot(x, y)</h2>

Kya hai?

Ek simple line chart banata hai X aur Y values ke beech.

Use Kab?

Jab aapko trend ya growth dikhani ho.

Example:

Time vs Temperature graph banana ho.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y)
plt.show()
```

<h2> plt.bar(x, y) → Bar Chart </h2>

```python
import matplotlib.pyplot as plt

days = ['Mon', 'Tue', 'Wed']   #X-Axis
sales = [100, 150, 120]        #Y-Axis

plt.bar(days, sales)
plt.show()
```

<h2> plt.scatter(x, y) → Scatter Plot </h2>

Kya hai?

Ek scatter plot banata hai — dots lagte hain har point pe.

Use Kab?

Jab aapko do values ke beech relation ya pattern dikhana ho.

```python
import matplotlib.pyplot as plt

x = [5, 7, 8, 7, 2]
y = [99, 86, 87, 88, 100]

plt.scatter(x, y)
plt.show()
```

<h2> plt.pie(data) </h2>

Kya hai?

Ek pie chart banata hai — data ke portions round chart mein.

Use Kab?

Jab aapko data ka percentage ya hissa dikhana ho.

Example:

Budget ka kitna part kis cheez mein kharcha hua

```python
import matplotlib.pyplot as plt

sizes = [40, 30, 20, 10]
labels = ['Rent', 'Food', 'Transport', 'Other']

plt.pie(sizes, labels=labels)
plt.show()
```

<h2> plt.xlabel("X Axis Label") </h2>

Kya hai?

Ye graph ke axis ka naam dikhate hain.

Kyun?

Taki dekhne wale ko samajh aaye x aur y kya represent kar rahe hain.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [10, 20, 30]

plt.plot(x, y)
plt.xlabel("Time (in hours)")
plt.show()
```

<h2> plt.ylabel("Y Axis Label") </h2>

```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [5, 15, 25]

plt.plot(x, y)
plt.ylabel("Speed (km/hr)")
plt.show()
```

<h2> plt.title("Your Title Here") </h2>

Kya hai?

Chart ka upar title dena.

Example:

"Daily Sales", "Population Growth", etc.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [10, 20, 30]

plt.plot(x, y)
plt.title("Growth Over Time")
plt.show()
```

<h2> plt.legend()</h2>

Kya hai?

Jab aap ek chart mein multiple lines ya bars dikhate ho, ye batata hai kaun si line kis cheez ko represent karti hai.

Use Kab?

Multiple datasets wale graph mein

```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y1 = [10, 20, 30]
y2 = [15, 25, 35]

plt.plot(x, y1, label='Product A')
plt.plot(x, y2, label='Product B')
plt.legend()
plt.show()
```

<h2> plt.grid(True)</h2>

Kya hai?

Graph ke peeche lines (grid) dikhata hai for better readability.

Layman:

"Graph ke paper pe lines khinch do – jisse point samajh aaye."

```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [2, 4, 6]

plt.plot(x, y)
plt.grid(True)
plt.show()
```

<h2> plt.savefig("filename.png")</h2>

Kya hai?

Graph ko image ke roop mein save karta hai.

Use Kab?

Jab report ya presentation ke liye chart chahiye ho

```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [3, 6, 9]

plt.plot(x, y)
plt.title("Saved Graph")
plt.savefig("graph.png")
```

<h2> Style Options (Color, Linestyle, Marker, etc.) </h2>

What is a Marker in Matplotlib?

A marker is a symbol that appears on each data point in a graph.

| Marker | Shape       | Use Example  |
| ------ | ----------- | ------------ |
| `'o'`  | Circle      | `marker='o'` |
| `'s'`  | Square      | `marker='s'` |
| `'^'`  | Triangle up | `marker='^'` |
| `'*'`  | Star        | `marker='*'` |
| `'x'`  | X mark      | `marker='x'` |
| `'D'`  | Diamond     | `marker='D'` |


```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y, color='red', linestyle='--', marker='s', label='Line1')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

models = ['M340i xDrive','Z4 M40i','X4 M40i','M2 Coupe',
          'M4 Competition','M5','M8 Competition','XM']
prices = [72.9, 90.9, 98.2, 103.0, 153.0, 199.0, 244.0, 260.0]

plt.figure(figsize=(12,6))
plt.bar(models, prices, color='navy')  # marker works only in plot lines
plt.title('BMW M‑Series Models vs Ex‑Showroom Price (India, in lakh ₹)')
plt.xlabel('Model')
plt.ylabel('Price (₹ lakh)')


plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Optional: show exact value labels on top
for i, v in enumerate(prices):
    plt.text(i, v + 5, f"{v:.1f}", ha='center')

#plt.tight_layout()
plt.savefig("BMW_Vs_Price.png")
plt.show()
```

ha stands for horizontal alignment.

| Value      | Meaning                      |
| ---------- | ---------------------------- |
| `'left'`   | Align text to the left of x  |
| `'center'` | Center text on x (default)   |
| `'right'`  | Align text to the right of x |


🔁 for i, v in enumerate(prices):
enumerate(prices) gives both:

i → the index of the current price (used to locate the x-axis position),

v → the actual price (the y-value).

🧾 plt.text(i, v + 5, f"{v:.1f}", ha='center')

This places a text label slightly above the top of the bar.

i → x-position of the bar (i.e. its index in the list).

v + 5 → y-position: a little above the top of the bar (adds 5 units above the bar height so text isn't inside the bar).

f"{v:.1f}" → formatted price label (1 decimal place, like 72.9).

ha='center' → center-aligns the text horizontally with the bar.
