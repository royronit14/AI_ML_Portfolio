Pie Chart

The autopct parameter in a pie chart is used to automatically label the slices with their percentage value.

The name stands for "automatic percentage". You give it a special formatting string that tells Matplotlib how you want the percentage to look.

How it Works
The most common format you'll see is autopct='%1.1f%%'. Let's break it down:

%: This tells Python that you're about to format a number.

1.1f: This specifies the number's format. It means a floating-point number with at least 1 digit before the decimal and exactly 1 digit after it.

%%: A single % is a special character, so you use two of them (%%) to tell Matplotlib to simply display a literal percent sign (%) on the chart.

To format a number means to change how it looks for display, without changing its actual mathematical value. It's all about presentation. ð¨

| Format String   | What it Does                                | Example Output |
|-----------------|------------------------------------------   |----------------|
| '%1.1f%%'       | Shows one decimal place. (Most common)      | 40.0%          |
| '%.2f%%'        | Shows two decimal places.                   | 40.00%         |
| '%1.0f%%'       | Shows zero decimal places (like an integer).| 40%            |
| '%d%%'          | Shows the value as a simple integer.        | 40%            |

```python
import matplotlib.pyplot as plt

# Sample Data
brands = ['Brand A', 'Brand B', 'Brand C', 'Others']
market_share = [40, 30, 15, 15]

# Create the plot
plt.pie(market_share, labels=brands, autopct='%1.1f%%') # autopct shows percentages

# Add title
plt.title('Smartphone Market Share')

# Show the plot
plt.show()
```
