<h1> SUBPLOTS </h1>

Subplots are multiple, separate plots that are arranged in a grid within a single figure. üñºÔ∏è

Think of a comic book page: the entire page is the figure, and each individual panel on that page is a subplot.

Why Use Subplots?
You use subplots primarily for two reasons:

Comparison: They are perfect for placing graphs side-by-side to easily compare different datasets or different views of the same data.

Organization: They allow you to group related visualizations together into one cohesive image, which is much cleaner than displaying many separate graphs.

<h2> Basic Way Using plt.subplot(nrows, ncol, index) </h2>

plt.subplot() is a function that creates and activates one single subplot at a time within a grid.

It's a simpler, older way to create subplots compared to plt.subplots(). Instead of creating all plots at once, you call plt.subplot() each time you want to switch to a different plot.

How it Works
You use it like this: plt.subplot(nrows, ncols, index)

<b>nrows:</b> The total number of rows in your imaginary grid.

<b>ncols:</b> The total number of columns in your imaginary grid.

<b>index:</b> The position of the subplot you want to create or activate. Important: The index starts at 1 in the upper-left corner and increases to the right.

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,15,25]

plt.subplot(1,2,1)  #1row, 2column, 1st subplot
plt.plot(x,y)
plt.title("Line Chart")

plt.subplot(1,2,2)  #1row, 2column, 2nd subplot
plt.bar(x,y)
plt.title("Bar Chart")

plt.tight_layout()
plt.show()
```

<h2> Professional Way </h2>

fig: This variable represents the entire figure. You use it for actions that affect the whole window, like saving the plot (fig.savefig()) or adding a main title (fig.suptitle()).

ax: This variable represents the plotting area of a subplot. You use it for almost all your plotting commands, like drawing lines (ax.plot()) or setting a subplot's title (ax.set_title()). If you create multiple subplots (e.g., plt.subplots(2, 2)), then ax will be an array of axes.

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,15,25]

#fig, ax = plt.subplots(nrows, ncols, figsize =[widhth, height])
fig, ax = plt.subplots(1,3, figsize = [10,5])

ax[0].plot(x,y)
ax[0].set_title("Line Chart")

ax[1].bar(x,y)
ax[1].set_title("Bar Chart")

ax[2].scatter(x,y)
ax[2].set_title("Hist Chart")
plt.tight_layout()
plt.show()
```
