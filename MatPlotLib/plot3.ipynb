<h3> 3Months Sale vs WeekDays </h3>

````python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
week1_sales = [10,8,9,12,22]
week2_sales = [12,9,5,12,28]

plt.xlabel("Days in a Week")
plt.ylabel("Sales")
plt.title("Sales per Week")
plt.plot(days, week1_sales, color = 'red', linestyle = '--', marker = "o", linewidth = 2, label = "week_1")
plt.plot(days, week2_sales, color = 'blue', linestyle = '--', marker = "o", linewidth = 2, label = "week_2")

plt.legend(loc = "lower right")
plt.show()
````

In simple terms, xticks() and yticks() are functions that let you get or set the properties of the ticks on the x-axis and y-axis, respectively. They give you direct control over where the tick marks appear and what they say.

What can you do with them?
You can use these functions to:

Set the Tick Positions: You can specify the exact data values where you want the major ticks to appear on an axis.

Set the Tick Labels: You can provide a custom list of labels for the ticks. This is useful for changing numbers to text (e.g., 1 to 'Jan', 2 to 'Feb') or for renaming categories.

Control Appearance: You can pass in other arguments to change properties like the rotation and font size of the tick labels.

````python
import matplotlib.pyplot as plt

# The positions on the axis where you want ticks
tick_positions = [0, 5, 10]

# The text you want to display at those positions
tick_labels = ['Low', 'Mid', 'High']

# Set the x-axis ticks and their labels
plt.xticks(ticks=tick_positions, labels=tick_labels, rotation=45)
plt.show()
````
